```python
import dash_bootstrap_components as dbc
import dash
from dash import Dash, dcc, html
from dash.dependencies import Input, Output, State

app = dash.Dash(external_stylesheets=[dbc.themes.QUARTZ])
SIDEBAR_STYLE = {
    "position": "fixed",
    "top": 0,
    "left": 0,
    "bottom": 0,
    "width": "16rem",
    "padding": "2rem 1rem"
}
CONTENT_STYLE = {
    "margin-left": "18rem",
    "margin-right": "2rem",
    "padding": "2rem 1rem",
}

nav_items = [
    dbc.NavItem(dbc.NavLink("Artist", id="page-1-link",href="#")),
    dbc.NavItem(dbc.NavLink("Lyrics", id="page-2-link",href="#")),
    dbc.NavItem(dbc.NavLink("Audio", id="page-3-link",href="#")),
]

page_1=html.Div([html.H3('Artist Analysis'),
                dcc.Graph(figure={
                    'data': [{
                        'x': [1, 2, 3],
                        'y': [3, 1, 2],
                        'type': 'bar'}]})])
page_2=html.Div([html.H3('Lyrics Analysis'),
                dcc.Graph(figure={
                    'data': [{
                        'x': [1, 2, 3],
                        'y': [3, 1, 2],
                        'type': 'scatter'}]})])
page_3=html.Div([html.H3('Audio Analysis'),
            dcc.Graph(figure={
                    'data': [{
                        'x': [1, 2, 3],
                        'y': [3, 1, 2],
                        'type': 'scatter'}]})])


sidebar = html.Div([   
        html.H3("Study Bird 551", className="display-4"),
        html.Hr(),
        html.P("Selection Bar", className="lead"),
        dbc.Nav(nav_items,
            vertical=True,
            pills=True)],style=SIDEBAR_STYLE)
timer=html.Div([
        html.Br(),
        dcc.RangeSlider(
            id='year-slider',
            min=2018,
            max=2022,
            value=[2018, 2019],
            step=1,
            marks={'2018':{'label':'2018','style': {'color': 'white'}},
                  '2019':{'label':'2019','style': {'color': 'white'}},
                  '2020':{'label':'2020','style': {'color': 'white'}},
                  '2021':{'label':'2021','style': {'color': 'white'}},
                  '2022':{'label':'2022','style': {'color': 'white'}}
                  }
    )])
content = html.Div([timer,
    html.Div(id='tabs-content-example-graph', children=page_1)],style=CONTENT_STYLE)

app.layout = html.Div([dcc.Location(id="url"), sidebar, content])


@app.callback(
    Output("tabs-content-example-graph", "children"), 
    [Input("page-1-link", "n_clicks"), 
     Input("page-2-link", "n_clicks"), 
     Input("page-3-link", "n_clicks")],
     State("tabs-content-example-graph", "children"))


    
def render_page_content(page_1_clicks, page_2_clicks, page_3_clicks, current_page):
    changed_id = [p['prop_id'] for p in dash.callback_context.triggered][0]
    if current_page == "This is page 1":
        page = page_1
    else:
        page = page_1
    if "page-1-link" in changed_id:
        return page_1
    elif "page-2-link" in changed_id:
        return page_2
    elif "page-3-link" in changed_id:
        return page_3
    else:
        return page
    
if __name__ == '__main__':
    app.run_server(debug=True, use_reloader=False)
```
