```python
import dash_bootstrap_components as dbc
from dash import Dash, dcc, html
from dash.dependencies import Input, Output

app = Dash(external_stylesheets=[dbc.themes.QUARTZ])
SIDEBAR_STYLE = {
    "position": "fixed",
    "top": 0,
    "left": 0,
    "bottom": 0,
    "width": "16rem",
    "padding": "2rem 1rem"
}
CONTENT_STYLE = {
    "margin-left": "18rem",
    "margin-right": "2rem",
    "padding": "2rem 1rem",
}
sidebar = html.Div([   
        html.H2("Sidebar", className="display-4"),
        html.Hr(),
        html.P("Year Selection", className="lead"),
        dbc.Nav([
                dbc.NavLink("2018", href="/page-1", active="exact"),
                dbc.NavLink("2019", href="/page-2", active="exact"),
                dbc.NavLink("2020", href="/page-3", active="exact"),
                dbc.NavLink("2021", href="/page-4", active="exact"),
                dbc.NavLink("2022", href="/page-5", active="exact")],
            vertical=True,
            pills=True)],style=SIDEBAR_STYLE)
tab = html.Div([
    html.H1('Dash Tabs component demo'),
    dcc.Tabs(id="tabs-example-graph", value='tab-1-example-graph', children=[
        dcc.Tab(label='Artist', value='tab-1-example-graph'),
        dcc.Tab(label='Lyrics', value='tab-2-example-graph'),
        dcc.Tab(label='Audio', value='tab-3-example-graph')
    ],colors={'border': '#d6d6d6', 'primary': '#1975FA', 'background': '#3c89bd'}),
    html.Div(id='tabs-content-example-graph')],style=CONTENT_STYLE)

app.layout = html.Div([dcc.Location(id="url"), sidebar, tab])

@app.callback(Output('tabs-content-example-graph', 'children'),
              Input('tabs-example-graph', 'value'))
def render_content(tab):
    if tab == 'tab-1-example-graph':
        return html.Div([
            html.H3('Tab content 1'),
            dcc.Graph(
                figure={
                    'data': [{
                        'x': [1, 2, 3],
                        'y': [3, 1, 2],
                        'type': 'bar'}]})])
    elif tab == 'tab-2-example-graph':
        return html.Div([
            html.H3('Tab content 2'),
            dcc.Graph(
                id='graph-2-tabs-dcc',
                figure={
                    'data': [{
                        'x': [1, 2, 3],
                        'y': [5, 10, 6],
                        'type': 'bar'}]})])
    elif tab == 'tab-3-example-graph':
        return html.Div([
            html.H3('Tab content 2'),
            dcc.Graph(
                id='graph-2-tabs-dcc',
                figure={
                    'data': [{
                        'x': [1, 2, 3],
                        'y': [5, 10, 6],
                        'type': 'bar'}]})])


if __name__ == '__main__':
    app.run_server(debug=True, use_reloader=False)
```
