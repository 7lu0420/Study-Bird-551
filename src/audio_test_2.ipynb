```python
import altair as alt
import datetime
import pandas as pd

year_input=[datetime.date(2019,1,1),datetime.date(2020,1,1)]
hits = pd.read_csv('../data/processed/audio_data_processed.csv', parse_dates=['Year'])
hits.drop("Unnamed: 0", axis=1, inplace=True)

hits['Rank_bin'] = pd.cut(hits['Rank'],bins=10,labels=['Top10','Top10-20','Top20-30','Top30-40','Top40-50','Top50-60','Top60-70','Top70-80','Top80-90','Top90-100'])

hits_c1 = hits[hits['Year'].isin(year_input)]
hits_c1_bin = hits_c1.groupby('Rank_bin').mean()
hits_c1_bin = hits_c1_bin.reset_index()
hits_c1_bin
```

```python
alt.Chart(hits_c1_bin[['Rank_bin','danceability','popularity']]).mark_bar().encode(
    x='Rank_bin',
    y='danceability:O',
    color='popularity:Q'
)
```
